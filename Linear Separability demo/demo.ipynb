```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron
```

```python
# Sample linearly separable data
X = np.array([
    [1, 2], [2, 3], [3, 3], [6, 5], [7, 8], [8, 8]
])

y = np.array([0, 0, 0, 1, 1, 1])
```

```python
# Train a perceptron to find a separating line
clf = Perceptron()
clf.fit(X, y)

# Get weights
w = clf.coef_[0]
b = clf.intercept_
```

```python
# Plotting
plt.figure(figsize=(6, 6))

# Scatter the points
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], color='red', label='Class 0')
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color='blue', label='Class 1')

# Draw the separating line: w1*x + w2*y + b = 0 -> y = -(w1*x + b)/w2
x_vals = np.linspace(0, 10, 100)
y_vals = -(w[0] * x_vals + b) / w[1]
plt.plot(x_vals, y_vals, 'k--', label='Separating Line')

plt.xlim(0, 10)
plt.ylim(0, 10)
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.grid(True)
plt.title('Linear Separability Demo')
plt.show()
```
