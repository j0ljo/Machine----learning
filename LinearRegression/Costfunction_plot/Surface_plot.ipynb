```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Data

X = np.array([1400, 1600, 1700, 1875, 1100, 1550, 2350, 2450, 1425, 1700])
y = np.array([245, 312, 279, 308, 199, 219, 405, 324, 319, 255])
n = len(X)
```

```python
# Create a meshgrid for m and b

m_vals = np.linspace(-0.1, 0.3, 100)
b_vals = np.linspace(0, 500, 100)
M, B = np.meshgrid(m_vals, b_vals)
```

```python
# Initialize cost surface
J = np.zeros_like(M)

# Compute cost for each (m, b)
for i in range(M.shape[0]):
    for j in range(M.shape[1]):
        m = M[i, j]
        b = B[i, j]
        predictions = m * X + b
        errors = y - predictions
        J[i, j] = np.mean(errors ** 2)
```

```python
# Plot 3D surface

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(M, B, J, cmap='viridis', edgecolor='none', alpha=0.9)

ax.set_xlabel('Slope (m)')
ax.set_ylabel('Intercept (b)')
ax.set_zlabel('Cost J(m, b)')
ax.set_title('Cost Function Landscape')

plt.tight_layout()
plt.show()
```
